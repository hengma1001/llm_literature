```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_pickle("./arxiv_hf.pkl")
df
```

```python
df['cat'] = [cat.split()[0].split('.')[0] for cat in df.categories]
```

```python
fig = px.scatter(
    df.iloc[::],
    x="umapE_0",
    y="umapE_1",
    color="cat",
    hover_data=["id", "title", "categories"],
    render_mode="webgl",
)
fig.update_traces(marker=dict(size=4))
fig.write_html("umap_hf_euc.html")
```

```python
fig = px.scatter(
    df.iloc[::],
    x="umapC_0",
    y="umapC_1",
    color="cat",
    hover_data=["id", "title", "categories"],
    render_mode="webgl",
)
fig.update_traces(marker=dict(size=4))
fig.write_html("umap_hf_cos.html")
```
