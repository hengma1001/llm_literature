```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

from scipy.spatial.distance import euclidean
```

```python
df = pd.read_pickle("./loc_hf.pkl")
```

```python
dist_mat = np.zeros([len(df), len(df)])
for i, emb0 in enumerate(df.embedding):
    emb0 = np.array(emb0)
    for j, emb1 in enumerate(df.embedding):
        emb1 = np.array(emb1)
        dist_mat[i, j] = euclidean(emb0, emb1) if i!= j else np.nan
```

```python
import plotly.express as px

fig = px.imshow(dist_mat, x=df.name, y=df.name, color_continuous_scale='RdBu')

fig.show()
fig.write_html("loc_euc.html")
```

```python
w_emb = df.iloc[-3].embedding

dist_mat = np.zeros([len(df), len(df)])
for i, emb0 in enumerate(df.embedding):
    emb0 = np.array(emb0)
    for j, emb1 in enumerate(df.embedding):
        emb1 = np.array(emb1)
        dist_mat[i, j] = euclidean(emb0, emb1, w=np.abs(w_emb)) if i!= j else np.nan

fig = px.imshow(dist_mat, x=df.name, y=df.name, color_continuous_scale='RdBu')

fig.show()
fig.write_html("loc_euc_weight.html")
```
